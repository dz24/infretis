```python
def write_bror(number, wait_time=7.5, write=False):
    """Lock a worker for wait_time seconds and write bror number."""
    # lock wait_time seconds
    current_time = time.time()
    worker = get_worker()
    while time.time()-wait_time < current_time:
        continue
        
    # write bror level
    bror_number = f'bror_{number}'
    # if write:
    with open(bror_number + '.txt', 'w', encoding='utf-8') as writer:
        writer.write(bror_number)
    # else:
    #     print(bror_number)
    # os. getcwd()
    
    return bror_number, worker.local_directory, os. getcwd(), current_time
```
